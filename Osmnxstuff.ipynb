```python
import networkx as nx
from pyvis.network import Network 
import osmnx as ox
import matplotlib.pyplot as plt

%matplotlib inline
ox.__version__
```

```python
# download/model a street network for some city then visualize it
G = ox.graph_from_place("Pinellas, Florida, USA", network_type="drive")
fig, ax = ox.plot_graph(G, node_size=0, edge_linewidth=0.5, edge_color="w")
```

```python
# get all building footprints in some neighborhood
place = "Saint Petersburg, Florida"
tags = {"building": True}
gdf = ox.geometries_from_place(place, tags)
gdf.shape
```

```python
# download/model a street network for some city then visualize it

G1 = ox.graph_from_place("Pinellas, Florida, USA", network_type="drive")
G2 = ox.graph_from_gdfs(gdf)
G = nx.compose(G1, G2)

fig, ax = ox.plot_graph(G, subgraphdata(), node_size=0, edge_linewidth=0.5, edge_color="w", figsize=(24, 24))


G.subgraph(subgraphdata())
```

```python
# get network of roads in county to be base of plot
G1 = ox.graph_from_place("Pinellas, Florida, USA", network_type="drive")

fig, ax = ox.plot_graph(G1, node_size=0, edge_linewidth=0.5, edge_color="w", figsize=(24, 24), bgcolor='k', show=False, close=False)

# add specific address represented as blue dot on screen
ax.scatter(-82.691796, 27.786347, color='b', zorder=3) #4225
ax.scatter(-82.7440749, 28.1023121, color='r', zorder=3) #3300
ax.scatter(-82.7025671, 27.8673034, color='y', zorder=3) #10751
ax.plot([-82.7440749, -82.7025671, -82.691796], [28.1023121, 27.8673034, 27.786347], 'C3', lw=2, zorder=2)
plt.show()
```
